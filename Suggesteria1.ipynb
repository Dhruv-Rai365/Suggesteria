We will select the useful columns like genres, movie_id, keywords, title, overview, cast, crew

```python
movies=movies[['movie_id','title','genres','keywords','overview','cast','crew']]
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
#convert wierd string into a list of words for genres
import ast
def convert_genres(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
movies['keywords'][0]
```

```python
#convert wierd string into a list of strings for keywords
def convert_keywords(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert_keywords)
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['crew'][0]
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['genres'][4]
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies=movies[["movie_id","title","tags"]]
```

```python
new_movies.head()
```

```python
new_movies['tags'][0]
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x :" ".join(x))
```

```python
new_movies.head()
```

```python
new_movies['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words ='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors[0]
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stemming)
```

```python
new_movies.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    siml_movie =similarity[movie_index]
    rec = sorted(list(enumerate(siml_movie)), reverse=True, key=lambda x:x[1])[1:6]
    for i in rec:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
